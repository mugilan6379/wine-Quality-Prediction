```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
wine_dataset = pd.read_csv('winequalityred.csv')
```

```python
wine_dataset.head(20)
```

```python
wine_dataset.isnull().sum()
```

```python
wine_dataset.describe()
```

```python
X = wine_dataset.drop(columns='quality',axis=1)
Y=wine_dataset['quality'].apply(lambda y:1 if y>6 else 0)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=2,test_size=0.2)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train,Y_train)
```

```python
Y_train_pred = classifier.predict(X_train)
Y_test_pred = classifier.predict(X_test)
train_acc = accuracy_score(Y_train_pred,Y_train)
test_acc = accuracy_score(Y_test_pred,Y_test)
print('Train accuracy: ',train_acc)
print('Test accuracy: ',test_acc)
```

```python
data = (8.0,0.59,0.16,1.8,0.065,3.0,16.0,0.9962,3.42,0.92,10.5)
array_data = np.asarray(data)
array_data_reshape = array_data.reshape(1,-1)
pred = classifier.predict(array_data_reshape)
print(pred)
if pred[0]==1:
    print('The wine is of GOOD quality')
else:
    print('The wine is of BAD quality')
```
